# Experiment 1.2
Extending the single-cell model from experiment 1.1 to multiple cells via a hypenetwork controlled by a vanilla VAE.


## 1.2.1
All cells in a single recording. A few (unrecorded) attempts at this were tried where the architecture was varied. The architecture that was settled on for this experiment was a rough duplicate of the model from 1.1, but where the ResNet blocks had their weights set from a pool of weights (hypernetwork). Each of the 3 CNN layers in a ResNet block had a pool of channel kernels 4x larger than the number of channels of the layer. These kernels were stored flat in a matrix termed a "warehouse". From the warehouse, the weights of a layer would be pulled. What weights was determined by an attention block controlled by the output of a VAE when encodes a learned embedding uniquely assigned to each recording-cell id pair. The VAE hidden dimension was set to 2. 2 was chosen to allow easy debugging; however, it is one of the suspected causes of performance being less than hoped for.

## 1.2.2
1.2.2 was repeated for all recordings. Results seemed promising, but inspecting the latent space showed very poor separation after 2 epoch. Training didn't proceed much longer as each epoch took about 12 hours to run.
