# Experiment 1.3
Spike prediction for multiple clusters, like in experiment 1.2, but instead of a VAE controlled hypernetwork, we simply concatenate the output of a VAE to layer 1 in the feadforward CNN model. 

## CNN model
Experiment 1.1.4 was done inbetween 1.2 and 1.3, and so the basic CNN model template used in 1.3 includes some of the changes of 1.1.4. This includes things like:

    - SE block
    - mean concatenation at the input
    - spike mask changed from "2" to "-1"


    


## 1.1.1 to 1.1.3
Not the first run, but the first recorded experiments. These three differ only in the setting of the $\beta$ value to $1x10^{-3}, 1x10^{-4}$ and $1x10^{-5}$ respectively. They shared a batch size (64), and z dimension (50). Within the first 2 million steps, there wasn't any obvious difference between their performance.




### 1.1.4
Here we modify 1.1.1 by doubling the l1 and l2 channels (I think l0 had 20 or so channels added also). The point was to check if the number of channels was a bottleneck. After 1 million steps, there isn't any obvious differences in performance.
